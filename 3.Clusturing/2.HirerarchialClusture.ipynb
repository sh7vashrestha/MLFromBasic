```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataSet = pd.read_csv("../Datas/10.Mall_Customers.csv")
X = dataSet.iloc[:, [3,4]].values
```

```python
for x in range(1,5):
    print(X[x])
```

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendogram for HClusturing")
plt.xlabel("customers")
plt.ylabel('Eucledian Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward' )
y_hc = hc.fit_predict(X)
```

```python
plt.scatter(X[y_hc==0, 0],X[y_hc==0, 1], s=100, c='red', label ='clusture1')
plt.scatter(X[y_hc==1, 0],X[y_hc==1, 1], s=100, c='blue', label ='clusture2')
plt.scatter(X[y_hc==2, 0],X[y_hc==2, 1], s=100, c='green', label ='clusture3')
plt.scatter(X[y_hc==3, 0],X[y_hc==3, 1], s=100, c='cyan', label ='clusture4')
plt.scatter(X[y_hc==4, 0],X[y_hc==4, 1], s=100, c='purple', label ='clusture4')
plt.title("HC-Cluster")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
plt.show()
```
