```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataSet = pd.read_csv("../Datas/4.Position_Salaries.csv")
x = dataSet.iloc[:,1:-1].values
y = dataSet.iloc[:,-1].values
```

```python
print(x)
```

```python
from sklearn.linear_model import LinearRegression
Lin_regressor = LinearRegression()
Lin_regressor.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
Lin_regressor_2 = LinearRegression()
Lin_regressor_2.fit(x_poly, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, Lin_regressor.predict(x), color = 'blue')
plt.title('Linear regressor')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
x_grid = np.arange(min(x), max(x),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, Lin_regressor_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Polynomial regressor')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
# For smoother curve
```
